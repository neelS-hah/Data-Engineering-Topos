```python
import requests
import pandas as pd  
import json 
import re
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population').text
soup = BeautifulSoup(website_url,'lxml')
my_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
links = My_table.findAll('tr')
row = []
diction = {}
count = 1 
for link in links:
    x = link.findAll('td')
    for elem in x:
        if elem.text[:-1] == str(count):
            count += 1
            diction[count] = row
            row = []
        row.append(elem.text[:-1])
diction.pop(2)
df = pd.DataFrame(diction)
df = df.T
df = df.drop(0, axis = 1).reset_index().drop('index', axis = 1)

#Creates a dataframe 

final_col = []
for elem in df[1]:
    if "[" in str(elem):
        x = re.search(r"\[([A-Za-z0-9_]+)\]", elem)
        if len(x.group(1)) == 1:
            final_col.append(elem[:-3])
        if len(x.group(1)) == 2:
            final_col.append(elem[:-4])
        continue
    final_col.append(elem)
df[1] = final_col
df
```

```python
df = df.rename(columns={1: "City", 2: "State", 3: "2018 Population", 4: "2008 population", 
                       5: "Percent Change", 6: "Land Area sq mi", 7: "Land Area sq km", 8:
                       "Population Density sq mi", 9: "Population Density sq km", 10: 
                        "Lat/Long degrees"})
df
```
